```python
import matplotlib.pyplot as plt
x = ["Part1","Part2","Part3","Part4","Part5"]
y = [98,56,67,95,88]
colors =["red","green","yellow","blue","orange"]
plt.bar(x,y,color = colors,edgecolor = "black")
plt.xlabel("Parts of Harry Potter",fontsize = 17,)
plt.ylabel("Populatity",fontsize = 17)
plt.title("Popualatity of Different parts of Harry Potter",fontsize = 17)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Part1","Part2","Part3","Part4","Part5"]
y = [98,56,67,45,88]
ex = [0,0,0,0.1,0]
plt.pie(y,labels = x,explode=ex,shadow = True,autopct = "%.2f",startangle = 180)
plt.savefig("pie.png",facecolor ="green",pad_inches = 0.4)
plt.show()
```
